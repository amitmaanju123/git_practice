```python
# A function to find the biggest of three numbers using nested if statements
def biggest(a, b, c):
    if a > b:
        if a > c:
            return a
        else:
            return c # c is big 
    else:
        if b > c:
           return b
        else:
            return c
a = int(input("Enter a value"))
b = int(input("Enter b value"))
c = int(input("Enter c value"))
# Function call
big = biggest(a, b, c)
print("big number= ", big)
```

# Variable positional arguments (*args)

```python
They make your function flexible, so it can take any number of arguments.

The *args syntax allows a function to accept any number of additional positional arguments. 
*args collects extra positional arguments into a tuple
```

```python
def add_numbers(*x):
    for i in x:
        
       print(i)
    # Calling the function and printing the result.
add_numbers(5, 20,30,40,45,4,56,7,2,3,4)
```

```python
def add(*nums):
    sum1=0
    for i in nums:
        
        sum1+=i
    return sum1
    
add(1,2,3,20)
```

```python
def add_all(*x):
    print(x)
    print(type(x))
add_all([1, 2, 3],4,5,6)   
```

```python
#sum of numbers
def add_all(*args):  #1,2,3
    total = 0
    for num in args: #1,2,3
        total += num
    return total

print(add_all(1, 2, 3))        # 6
print(add_all(5, 10, 15, 20))  # 50


#args is a tuple like (1, 2, 3)
```

```python
def calculate_sum(start_value,last_value,*numbers): #positional+*args
    print(start_value)
    print(last_value)
    total = start_value
    for num in numbers:
        total += num
    return total

# Pass multiple additional arguments.
print(calculate_sum( 1,1,1, 2, 3))
# Output: 16

# Pass no additional arguments.
print(calculate_sum(10,20))
# Output: 10
```

```python
In Python, a variable's scope determines where it is visible and can be accessed or modified. 
When a variable is referenced, Python searches for it in the following order:
Local (L): The current function's scope.
Enclosing (E): The scope of any nested functions.
Global (G): The top-level scope of a module or script.
Built-in (B): The special scope for Python's pre-defined names, like print() and len(). 
```

```python
Local scope
A variable defined inside a function has a local scope.
It exists only for the duration of the function call and 
is not accessible from outside that function. 
Each call to the same function creates a new local scope. 
```

```python
def my_function():
    x = 10  # `x` has a local scope
    print(x) # This will print 10

my_function()
# Output: 10

print(x)
# Output: NameError: name 'x' is not defined
```

```python
y=20
def my_function():
    x = 10  # `x` has a local scope
    return x # This will print 10

print(my_function())

# Output: 10
print(y)
## Output: NameError: name 'x' is not defined
```

```python
Enclosing scope
This scope exists for nested functions, 
where one function is defined inside another.
The inner function can access variables from the outer (enclosing) function's scope. 
This is used to create closures, where the inner function "remembers" the data from the outer scope.
```

```python
def outer_function():
    y = 20  # `y` is in the enclosing scope of `inner_function`
    def inner_function():
        print(y)  # Can access `y` from the enclosing scope
    return inner_function #The program calls the inner_function.

outer_function()

# Output: 20
```

```python
Explanation:

When outer_function() is called, it creates x = 10 and defines inner_function.

It returns the inner_function without calling it yet.

my_func now ‚Äúremembers‚Äù the value of x even though outer_function has finished.

When you call my_func(), it still prints x = 10.
```

```python
#It‚Äôs defined inside outer_function, 
#so it can ‚Äúsee‚Äù variables from outer_function‚Äôs scope.
```

```python
def outer_function():
    x = 10  # Variable in outer scope

    def inner_function():
        print("x =", x)  # Accesses `x` from the enclosing scope

    return inner_function # Return the inner function itself

# Get the inner function (closure)
outer_function()

# Now call it
```

```python
Defining outer_function:
Python creates the function, but doesn‚Äôt execute it yet.

Calling outer_function():

Inside outer_function, x = 10 is created.

Then inner_function is defined and returned.

However, you don‚Äôt store or call the returned function.
                     outer_function()
only creates and returns the inner function ‚Äî it does not execute it.

That means nothing is printed yet.
```

```python
Global scope
A variable defined at the top level of a script or module has a global scope
and is accessible throughout the program
```

```python
z = 40  # `z` has a global scope

def my_function():
    z +=1  # Can access `z` from the global scope
    print(z)
my_function()
# Output: 40
#print(z)
# Output: 40
```

```python
z = 40  # `z` has a global scope

def my_function():
    global z
    z += 1  # Can access `z` from the global scope
    print(z)
my_function()
# Output: 40
#print(z)
# Output: 40
```

```python
count = 0  # Global variable

def increment():
    global count  # Tell Python to use the global variable
    count += 1
    print("Inside function:", count)

increment()
increment() #The global count keeps its updated value between function calls.
print("Outside function:", count)
```

```python
Built-in scope
This is the broadest scope, containing all of Python's pre-defined names,
like built-in functions (print, len) and exceptions. 

This scope is loaded when the Python interpreter starts and is always available. 

It is generally bad practice to override or "shadow" these built-in names with your own variables.  
```

```python
print("This uses the built-in print function")
# Output: This uses the built-in print function

# Avoid doing this:
len = 5
print(len("hello")) 
```
